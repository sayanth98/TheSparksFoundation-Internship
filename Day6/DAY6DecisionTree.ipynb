# Prediction using Decision Tree Algorithm

Decision tree is the most powerful and popular tool for classification and prediction. A Decision tree is a flowchart like tree structure, where each internal node denotes a test on an attribute, each branch represents an outcome of the test, and each leaf node (terminal node) holds a class label

### Author : Sayanth R Nath

```python
#Importing the libraries 
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
data = pd.read_csv('Iris.csv',encoding='utf-8')
data.head()
```

```python
df=data
```

```python
from sklearn import tree
```

```python
A=tree.DecisionTreeClassifier(criterion='entropy', splitter='best')
```

#### Now we can store columns 2-5 as X and last column as y 

```python
X=df.iloc[:,1:5]
y=df['Species']
```

### We can now divide this into train-test splits!

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

#### Fitting the training data and predicting test data

```python
B=A.fit(X_train,y_train)
```

```python
y_pred=B.predict(X_test)
y_pred
```

### Classification Report and Accuracy Score

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

### Confusion Matrix

```python
print(confusion_matrix(y_pred,y_test))
```

```python
import seaborn as sns
sns.heatmap(confusion_matrix(y_pred,y_test),annot=True)
```

```python
col_names=list(df.columns.values)
```

## Drawing the Tree

```python
plt.figure(figsize=(25,13))
tree.plot_tree(A.fit(X,y),feature_names=col_names,filled=True, precision=3, proportion=True,rounded=True)
plt.show()
```
