## Statistic Figures

We can now apply 5 functions count,min,max,sum,standard deviation and mean 

```python
print("Count:",df["Sales"].count())
print("Mean:",df["Sales"].mean())
print("min:",df["Sales"].min())
print("max:",df["Sales"].max())
print("std:",df["Sales"].std())
```

```python
print("Count:",df["Quantity"].count())
print("Mean:",df["Quantity"].mean())
print("min:",df["Quantity"].min())
print("max:",df["Quantity"].max())
print("std:",df["Quantity"].std())
```

```python
print("Count:",df["Discount"].count())
print("Mean:",df["Discount"].mean())
print("min:",df["Discount"].min())
print("max:",df["Discount"].max())
print("std:",df["Discount"].std())
```

```python
print("Count:",df["Profit"].count())
print("Mean:",df["Profit"].mean())
print("min:",df["Profit"].min())
print("max:",df["Profit"].max())
print("std:",df["Profit"].std())
```

```python
df.index
```

```python
df.pivot_table(index='State')
```

### We can infer that the following states - " Arizona,Colorado,Florida,Illinois,North Carolina,Ohio,Pennsylvania,Oregon,Texas,Tennessee " have a net loss and we can see from the pivot table it is mainly because all these states have a  discount rates of almost 30% or above.

### Also the following states - " Alabama,Arkansas,District of Columbia,Georgia , Indiana ,Michigan,Minnesota,Missouri,Rhode Island,Vermont,Wyoming " have made a huge profit because all these states have almost zero discount rates.

### We can conclude that in order to maximize the profit , the discount rates in relevant states should be reduced.

```python
df.pivot_table(index='Category')
```

### From the above pivot table , we can infer that Furniture and Office supplies have made a very less profit when compared to 'Technology' with almost similar values of Quantity and Discount.

```python
df.pivot_table(index=['Category','Sub-Category'])
```

### "Book cases,Tables and Supplies" have made net loss whereas Copiers made a huge profit 
### Copiers and Machines had better sales when compared to other categories

```python
df.pivot_table(index='Ship Mode')
```

### Ship mode is not a major factor in deciding Discount , Profit , Sales & Quantity

```python
df.pivot_table(index='Region')
```

### Central areas had larger Discount rates while East and West Regions made high profits and South region topped the sales charts

```python
sns.set_style('whitegrid')
sns.countplot(x='Region',data=df,palette='deep',edgecolor=sns.color_palette('dark',8))
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Ship Mode',data=df,palette='deep',edgecolor=sns.color_palette('dark',8))
```

```python
region_hist = sns.FacetGrid(df, col='Ship Mode', palette='rainbow')
region_hist.map(plt.hist, 'Region')
region_hist.set_ylabels('Number')
```

```python
region_hist = sns.FacetGrid(df, col='Category',row='Sub-Category', palette='rainbow')
region_hist.map(plt.hist, 'Region')
region_hist.set_ylabels('Number')
```
