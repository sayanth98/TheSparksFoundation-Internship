# Exploratory Data Analysis - Terrorism

Exploratory data analysis is an approach to analyzing data sets to summarize their main characteristics, often with visual methods. A statistical model can be used or not, but primarily EDA is for seeing what the data can tell us beyond the formal modeling or hypothesis testing task. 

### Author : Sayanth R Nath

```python
#Importing the libraries 
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
data = pd.read_csv('globalterrorismdb_0718dist.csv',encoding='latin1')
data.head()
```

```python
data.describe(include='all')
```

```python
pd.set_option('display.max_columns', 999)
```

```python
data.head()
```

```python
df=data
```

```python
df.shape
```

## Dropping all columns with NaN value 

```python
df=df.dropna(axis=1)
```

### 106 columns deleted

```python
df.head(5)
```

```python
country_count=df['country_txt'].value_counts()
```

```python
country_counts=country_count.to_frame()
```

### Creating a new dataframe in descending order of No. of occurences of terrorist attacks !

```python
country_counts=country_counts.reset_index()
```

```python
country_counts.rename(columns={'index':'Country','country_txt':'Occurences'},inplace=True)
```

```python
country_counts
```

### We can infer that the countries Iraq,Pakistan,Afghanistan and India had more than 10,000 occurences of terrorist attacks and thus they are more prone to terrorism and can be considered as the hot zones of terrorism in the world . 

```python
region_count=df['region_txt'].value_counts()
```

```python
region_counts=region_count.to_frame()
```

```python
region_counts=region_counts.reset_index()
```

```python
region_counts.rename(columns={'index':'Area','region_txt':'Occurences'},inplace=True)
```

```python
region_counts
```

```python
df['region_txt'].value_counts().plot(kind='bar')
```

## From the above graph , we can infer that Middle East & North Africa & South Asia are the regions with a high terrorist attack rate

```python
df['region_txt'].value_counts().plot(kind='pie',figsize=(8, 8))
```

### This is a pie chart depicting the same.

# Data Analysis

```python
df.head(5)
```

### Since all the above columns are not required , we have selected only the following columns - ['eventid' , 'iyear' , 'country' , 'country_txt', 'region' , 'region_txt' , 'attacktype1', 'attacktype_txt', 'targtype1','targtype1_txt','weaptype1','weaptype1_txt']

### dataset is thus reduced to 12 columns 

```python
df=df[['eventid' , 'iyear' , 'country' , 'country_txt', 'region' , 'region_txt' , 'attacktype1', 'attacktype1_txt', 'targtype1','targtype1_txt','weaptype1','weaptype1_txt']]
```

```python
df
```

```python
attacktype_count=df['attacktype1_txt'].value_counts()
attacktype_counts=attacktype_count.to_frame()
attacktype_counts=attacktype_counts.reset_index()
attacktype_counts.rename(columns={'index':'Attack Type','attacktype1_txt':'Occurences'},inplace=True)
```

```python
attacktype_counts
```

```python
df['attacktype1_txt'].value_counts().plot(kind='bar')
```

### Bombing & Armed assault killed the most people while least number of people where killed in hijacking and unarmed assault and hostage situations

```python
targtype_count=df['targtype1_txt'].value_counts()
targtype_counts=targtype_count.to_frame()
targtype_counts=targtype_counts.reset_index()
targtype_counts.rename(columns={'index':'Target Type','targtype1_txt':'Occurences'},inplace=True)
```

```python
targtype_counts
```

```python
df['targtype1_txt'].value_counts().plot(kind='bar')
```

### Private citizens & Property were targetted more frequently than Military,Police & Government officials  

```python
weapon_count=df['weaptype1_txt'].value_counts()
weapon_counts=weapon_count.to_frame()
weapon_counts=weapon_counts.reset_index()
weapon_counts.rename(columns={'index':'Weapon Type','weaptype1_txt':'Occurences'},inplace=True)
```

```python
weapon_counts
```

```python
df['weaptype1_txt'].value_counts().plot(kind='bar')
```

## Explosives and Firearms were the most commonly used weapons while all other weapons were used scarcely
